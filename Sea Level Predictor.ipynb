```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
def draw_plot():
        # Read data from file
        df=pd.read_csv(r"C:\Users\Elias\Downloads\epa-sea-level - Copy.csv")
   
        # Create scatter plot
        fig = plt.subplots()
        plt.scatter(x='Year',y='CSIRO Adjusted Sea Level',data=df)
        plt.xlabel('Year')
        plt.ylabel('CSIRO Adjusted Sea Level')  
    
        # Create first line of best fit
        slope,intercept,r,p,se = linregress(x=df['Year'],y=df['CSIRO Adjusted Sea Level'])
        x=range(1880,2051)
        y=slope*x + intercept
        plt.plot(x,y,label='line of best fit')
    
       # Create second line of best fit
        df1= df[df['Year']>=2000]
        slope1,intercept1,r1,p1,se1 = linregress(x=df1['Year'],y=df1['CSIRO Adjusted Sea Level'])
        X=range(2000,2051)
        Y=slope1*X + intercept1
        plt.plot(X,Y,label='line of best fit2')
    
      # Add labels and title
        plt.title('RIse in Sea Level ')
        plt.xlabel('Year')
        plt.ylabel('Sea Level (Inches)')
    
      # Save plot and return data for testing (DO NOT MODIFY)
        plt.savefig('sea_level_plot.png')
        return plt.gca()
 
 
```

```python
draw_plot()
```
